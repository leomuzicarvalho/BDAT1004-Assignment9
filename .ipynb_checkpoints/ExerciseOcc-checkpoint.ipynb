### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep='|')
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby(['occupation']).mean().drop('user_id',1)
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users.rename(columns={'user_id': 'count'}).loc[users['gender']=='M'].groupby('occupation').count().drop(columns=['age','gender','zip_code']).sort_values('count', ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
users.groupby('occupation').max().drop(columns=['user_id','gender','zip_code']).sort_values('age', ascending=False).head()
```

```python
users.groupby('occupation').min().drop(columns=['user_id','gender','zip_code']).sort_values('age', ascending=True).head()
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).mean().drop(columns=['user_id'])
```

### Step 8.  For each occupation present the percentage of women and men

```python
newDF = users.groupby(['occupation', 'gender']).agg({'user_id': 'sum'})

newDF.groupby(level=0).apply(lambda x:100 * x / float(x.sum()))
```
